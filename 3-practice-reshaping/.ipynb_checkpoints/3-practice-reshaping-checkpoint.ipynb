# 6

```python
medals = medals.merge(
    drinks[['country','continent']],
    left_on='Country',
    right_on='country',
    how='left'
).drop(columns='country')

#drop 'country' เพราะมันติดไปด้วย

medals.head(15)
```

# 7

```python
medals.loc[medals.continent.isna(), 'Country'].to_frame(name='Country with NaN continent')
```

# 8

```python
def getContinent(country: str):
    country = country.lower()
    if any(i in country for i in ["australasia"]):
        return 'OC'
    elif any(i in country for i in ["canada", "united states", "mexico"]):
        return 'NA'
    elif any(i in country for i in ["cuba", "jamaica"]):
        return 'SA'
    elif any(i in country for i in ["czech ", "yugoslavia", "germany", "great britain", "russia", "soviet"]):
        return 'EU'

    return 'OT'

medals['continent'] = medals['continent'].fillna(medals.Country.apply(getContinent))

#Wide table full
medals_wide = medals.copy()

medals.head(20)
```

# 9

```python
medals = medals.set_index(['continent', 'Country']).drop('Code',axis=1)
medals
```

# 10

```python
medals.sort_index(ascending=[False,True], inplace=True)
medals_without_stack = medals.copy
medals
```

# 11

```python
x = pd.MultiIndex.from_tuples([
    ("Summer", "SummerGame"),
    ("Summer", "SummerGold"),
    ("Summer", "SummerSilver"),
    ("Summer", "SummerBronze"),
    ("Winter", "WinterGame"),
    ("Winter", "WinterGold"),
    ("Winter", "WinterSilver"),
    ("Winter", "WinterBronze")    
])
medals.columns = x
medals
```

# 12

```python
medals = medals.swaplevel(axis=1)
medals
```

# 13

```python
medals_wide.drop('Code', axis=1, inplace=True)
medals_wide.head()
```

```python
medals_long = medals_wide.melt(
    id_vars=['continent', 'Country'],
    value_vars=[i for i in medals_wide.columns if i not in ['continent','Country']],
    var_name='SeasonMedal',
    value_name='Count'
)

medals_long[['Season', 'Medal']] = medals_long['SeasonMedal'].str.extract(r'(Summer|Winter)(Game|Gold|Silver|Bronze)')
medals_long = medals_long[['continent', 'Country', 'Season', 'Medal', 'Count']]
medals_long.head(20)
```

# 14

```python
medals_long[(medals_long['Medal'] == 'Gold') & (medals_long['Season'] == 'Summer')]\
.groupby('continent')['Count']\
.sum()\
.sort_values(ascending=False)\
.to_frame(name='Sumary of Summer Gold')
```

# 15

```python
medals_long.groupby('continent')\
.size()\
.to_frame(name='Amount of countries')
```

# 17

```python
df17 = medals_long.pivot_table(
    index='Medal',
    columns='Season',
    values='Count',
    aggfunc='sum',
    fill_value=0
).drop('Game')

df17
```

# 18

```python
medals_long[(medals_long.Season == 'Summer') & (medals_long.Medal == 'Gold')]\
.groupby('continent')['Count']\
.mean()\
.to_frame(name='Average Gold Summer Medal')
```

# 19

```python
summer_data = medals_long[medals_long['Season'] == 'Summer']
winter_data = medals_long[medals_long['Season'] == 'Winter']

summer_count = summer_data.groupby(['continent'])['Country'].nunique()
winter_count = winter_data.groupby(['continent'])['Country'].nunique()

olympic_participation = pd.DataFrame({
    'Summer': summer_count,
    'Winter': winter_count
})

olympic_participation
```
