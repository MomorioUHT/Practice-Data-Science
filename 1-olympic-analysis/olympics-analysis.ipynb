# 27

```python
import numpy as np

TotalScoreWinter = (3 * (medals.WinterGold)) + (2 * (medals.WinterSilver)) + (1 * (medals.WinterBronze))
TotalScoreSummer = (3 * (medals.SummerGold)) + (2 * (medals.SummerSilver)) + (1 * (medals.SummerBronze))

WinterScorePerGame = np.where(medals.WinterGame > 0, TotalScoreWinter / medals.WinterGame, 0)
SummerScorePerGame = np.where(medals.SummerGame > 0, TotalScoreSummer / medals.SummerGame, 0)

medals["perGame"] = WinterScorePerGame + SummerScorePerGame
df27 = medals.sort_values("perGame", ascending=True)
df27.head(10)
```
