# 10

```python
five_min_mean = df.Magnitude.resample('5min').mean()

indexMax = five_min_mean.idxmax()
valueMax = five_min_mean.max()

print(f"Maximum Mean = {valueMax}")
print(f"Max Mean Time = {indexMax}")
```

# 12

```python
df['Time Difference'] = df.index.diff()

end = df['Time Difference'].idxmax()
diff = df.index.diff().max() #From 11
start = end-diff

print(f"Start = {start}")
print(f"End = {end}")
```

# 13

```python
df['Prev Difference'] = abs(df['Magnitude'] - df['Magnitude'].shift(-1))
df['Next Difference'] = abs(df['Magnitude'].shift(1) - df['Magnitude'])

df['Strong'] = (df['Prev Difference'] > 2) & (df['Next Difference'] > 2)
df13 = df.loc[df['Strong']].copy()
df13
```

# 14

```python
per_hour = df.resample('h').size()
df13['hour_count'] = df13.index.floor('h').map(per_hour)
df13
```
